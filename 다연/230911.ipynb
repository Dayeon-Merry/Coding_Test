# 1. 수박수박수

```python
def solution(n):
    answer = ''
    for i in range(n):
        if i % 2 == 0:
            answer = answer + '수'
        elif i % 2 == 1:
            answer = answer + '박'
    return answer
```

```python
solution(10)
```

# 2. 내적

```python
def solution(a, b):
    sum = 0
    for i in range(len(a)):
      sum += a[i]*b[i]
    return sum
```

```python
li1 = [1,2,3,4]
li2 = [-3,-1,0,2]
solution(li1, li2)
```

# 3. 부족한 금액 계산하기

```python
# price와 count를 곱한 값이 money를 넘어가면 넘어가는만큼 return, 넘어가지 않으면 0을 return
def solution(price, money, count):
    result = 0
    num = sum([price*c for c in range(count+1)])
    if num > money:
      result = abs(money - num)
    return result
```

```python
solution(3, 20, 4)
```

# 4. 문자열 섞기

```python
# str1과 str2는 알파벳 소문자로 이루어진 길이가 같은 문자열
def solution(str1, str2):
    result = ''
    for i in range(len(str1)):
      result = result + str1[i] + str2[i]
    return result
```

```python
str1 = 'aaaa'
str2 = 'bbbb'
solution(str1, str2)
```

# 5. 수열과 구간 쿼리 4

```python
def solution(arr, queries):
    for query in queries:
      idxs = list(filter(lambda n: n%query[-1]==0, list(range(query[0], query[1]+1))))
      for idx in idxs:
        arr[idx] += 1
    return arr
```

```python
arr = [0, 1, 2, 4, 3]
queries = [[0, 4, 1],[0, 3, 2],[0, 3, 3]]
solution(arr, queries)
```
